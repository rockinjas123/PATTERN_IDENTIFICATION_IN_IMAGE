```python
# importimg libraries
import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the main image
image=cv2.cvtColor(cv2.imread("WaldoBeach.jpg"),cv2.COLOR_BGR2RGB)
plt.imshow(image,cmap="gray")
plt.title("Main image")
```

```python
# converting to grayscale
image_gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray,cmap="gray")
plt.title("Main grayscaled image")
```

```python
# reading the pattern to be detcted
pattern=cv2.cvtColor(cv2.imread("waldo.jpg"),cv2.COLOR_BGR2RGB)
plt.imshow(pattern,cmap="gray")
plt.title("Pattern to be detected")
```

```python
pattern_gray=cv2.cvtColor(pattern,cv2.COLOR_RGB2GRAY)
plt.imshow(pattern_gray,cmap="gray")
plt.title("Pattern to be detected in grayscale")
```

```python
result=cv2.matchTemplate(image_gray,pattern_gray,cv2.TM_CCOEFF)
```

```python
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(result)
```

```python
#creating the box
top_left = max_loc
bottom_right=(top_left[0]+50,top_left[1]+50)
cv2.rectangle(image,top_left,bottom_right,(0,0,255),5)
```
